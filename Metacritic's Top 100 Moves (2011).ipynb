```python
#================================ Part 1 - Contacting Metacritic ================================
link = 'https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=2011&sort=desc&view=detailed'

#Python request pool
http = urllib3.PoolManager(ca_certs = certifi.where())

#Initiating web request
sesh = requests.Session()
sesh.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'
r = sesh.get(link)
results = r.content

#Converting raw bytes to strings
datastring = str(results, "utf-8")
```

```python
#================================ Part 2 - Retrieving Contents w/ Regex ================================
#Regex to web-scrape Movie Title
regex = '<a href=.*class="title"><h3>(.*)<\/h3>'
pattern = re.compile(regex)
title = re.findall(pattern, datastring)

#Regex to web-scrape Release Date
regex = '<div class="clamp-details">\s*<span>(.*)\<'
pattern = re.compile(regex)
release = re.findall(pattern, datastring)

#Regex to web-scrape Movie Description
regex = '<div class="summary">\s(.*)'
pattern = re.compile(regex)
desc = re.findall(pattern, datastring)

#Regex to web-scrape Metascore
regex = '<div class="clamp-score-wrap">\s*.*\s<div class="metascore_w large movie positive">(\d\d)'
pattern = re.compile(regex)
score = re.findall(pattern, datastring)

#Regex to web-scrape Thumbnail Photo URL
regex = '<a href=".*<img\ssrc="(https.*)" alt'
pattern = re.compile(regex)
img = re.findall(pattern, datastring)
```

```python
#================================ Part 3 - Converting Data into a DataFrame & Printing ================================
#Creating a dataframe that displays the movies in a neat table
df = pd.DataFrame([title, release, desc, score, img], index = (['Title', 'Release', 'Desc', 'Score', 'Img URL']) )

#Transposing to re-orient the table (changing the rows (title, release, etc) to columns)
tdf = df.transpose()
print('Top 100 Movies 2011 (Metacritic)')
tdf
```
